```python
import pandas as pd
```

```python
df = pd.read_csv('data_bencana_2018-2022.csv', delimiter = ';')
```

```python
# Define the new column names
new_column_names = [
    'No',
    'Kode Identitas Bencana',
    'ID Kabupaten',
    'Tanggal Kejadian',
    'Kejadian',
    'Lokasi',
    'Kabupaten',
    'Provinsi',
    'Kronologi & Dokumentasi',
    'Penyebab',
    'Meninggal',
    'Hilang',
    'Terluka',
    'Rumah Rusak',
    'Rumah Terendam',
    'Fasum Rusak'
]

# Rename the columns
df.columns = new_column_names
```

```python
df = df.drop(index = [0], axis = 0)
```

```python
df
```

```python
df['Meninggal'] = df['Meninggal'].fillna(0)
df['Terluka'] = df['Terluka'].fillna(0)
df['Hilang'] = df['Hilang'].fillna(0)
df['Rumah Terendam'] = df['Rumah Terendam'].fillna(0)
```

```python
df.drop('Kronologi & Dokumentasi', axis = 1)
```

```python
df['Penyebab'] = df['Penyebab'].fillna('Belum Diketahui')
```

```python
df['Tanggal Kejadian'] = pd.to_datetime(df['Tanggal Kejadian'])
```

```python
filtered_data = df[(df['Kejadian'] == 'BANJIR') & (df['Lokasi'].str.contains('Bekasi')) & (df['Tanggal Kejadian'].dt.year == 2022)]
filtered_data
```
