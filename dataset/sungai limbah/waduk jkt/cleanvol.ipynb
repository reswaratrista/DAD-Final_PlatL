```python
import pandas as pd
```

```python
data = pd.read_csv('volume_gabung.csv')
```

```python
data = data.groupby(['tahun', 'bulan'])['volume_sampah_perhari(m3)'].sum().reset_index()
data
```

```python
data['bulan'] = data['bulan'].astype(str)
data['tahun'] = data['tahun'].astype(str)
```

```python
data['tanggal'] = '1/' + data['bulan'] + '/' + data['tahun']
```

```python
data['tanggal'] = pd.to_datetime(data['tanggal'], format='%d/%m/%Y')
```

```python
data = data.sort_values(by='tanggal')
```

```python
data['tanggal'] = data['tanggal'].dt.strftime('%B %Y')
```

```python
data
```

```python
data = data.drop(['bulan', 'tahun'], axis=1)
```

```python
data.to_csv('volume_gabung_sortedd.csv', index=False)
```
