```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wilayah_datasets = {}
file_csv_wilayah = {
    'jawa barat': 'dataset\perkiraan cuaca per kecamatan\kecamatanforecast-jawabarat.csv',
    'jawa timur': 'dataset\perkiraan cuaca per kecamatan\kecamatanforecast-jawatimur.csv',
    'jawa tengah': 'dataset\perkiraan cuaca per kecamatan\kecamatanforecast-jawatengah.csv',
    'aceh': 'dataset\perkiraan cuaca per kecamatan\kecamatanforecast-aceh.csv',
    'papua barat': 'dataset\perkiraan cuaca per kecamatan\kecamatanforecast-papuabarat.csv',
    'papua selatan': 'dataset\perkiraan cuaca per kecamatan\kecamatanforecast-papua.csv'
}
```

```python
def clean_data(data_csv):
    data = pd.read_csv(data_csv, delimiter = ';')

    new_column_names = [
        'ID Lokasi',
        'Waktu',
        'Suhu Udara Minimum',
        'Suhu Udara Maksimum',
        'Kelembapan Udara Minimum',
        'Kelembapan Udara Maksimum',
        'Kelembapan Udara (%)',
        'Suhu Udara (C)',
        'Kode cuaca',
        'Arah Angin',
        'Kecepatan Angin (km/jam)'
    ]

    data.columns = new_column_names

    data = data.drop(columns=['Suhu Udara Minimum', 'Suhu Udara Maksimum', 'Kelembapan Udara Minimum', 'Kelembapan Udara Maksimum'])
    
    data['Waktu'] = pd.to_datetime(data['Waktu'], format='%Y-%m-%d %H:%M:%S')

    data['Tanggal'] = data['Waktu'].dt.date

    return data
```

```python
for wilayah, nama_file in file_csv_wilayah.items():
    dataset = clean_data(nama_file)
    wilayah_datasets[wilayah] = dataset
```
